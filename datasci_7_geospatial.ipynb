```python
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

## Dataset 1: Birth Statistics

```python
df1 = gpd.read_file('Birth_Statistics.geojson')
df1
```

```python
# shape of dataset
df1.shape
```

```python
# looking for missing values
missing_values = df1.isnull().sum()
missing_values
```

```python
# checking column names
df1.columns
```

```python
# map visualization
df1.plot("TeenBirth", legend=True)
```

```python
# another map visualization
fig, ax = plt.subplots(1, 1)

divider = make_axes_locatable(ax)

cax = divider.append_axes("bottom", size="5%", pad=0.1)

df1.plot(
    column="Birth_Rate",
    cmap='magma',
    ax=ax,
    legend=True,
    cax=cax,
    legend_kwds={"label": "Birth Rate", "orientation": "horizontal"},
);
```

### Insights:


This dataset contains the birth rates across Lake County, Illinois by zip code. The field attributes in this dataset include: low birth weight, Preterm birth, Teen birth, birth rate, and 1st Trimester of care.

1st map visualization: The rate of teen pregnancies appear to be low in this region except for one area which is located to the right of the county. There is also an area on the top right which appears to have a higher rate compared to other areas.

2nd map visualization: One pattern noted is that there is a higher percentage of doctor's visits in the southern region of Lake County. The northeast region of Lake County had around 60-70% of doctor's visits. The northwest region had approximately 72-80% and the southern region had around 80-85% of doctor's visits/1st trimester care.


## Dataset 2: Mortality Rates

```python
df2 = gpd.read_file('Mortality_Rates.geojson')
df2
```

```python
# checking for missing values
missing_values = df2.isnull().sum()
missing_values
```

```python
# map visualization
fig, ax = plt.subplots(1, 1)

divider = make_axes_locatable(ax)

cax = divider.append_axes("bottom", size="5%", pad=0.1)

df2.plot(
    column="Cancer_D",
    cmap='GnBu',
    ax=ax,
    legend=True,
    cax=cax,
    legend_kwds={"label": "Cancer Related Deaths (rate per 100,000)", "orientation": "horizontal"},
);
```

### Insights:

This dataset contains information about mortality rates in Lake County, Illnois. The field attributes contained in this dataset include: average age of death, cancer deaths, heart disease related deaths, and COPD related deaths. For the map visualization, I focused on cancer related deaths. This attribute shows the number of individuals (at a rate per 100,000) who have died of cancer as the underlying cause.

One anomaly in the Cancer related deaths (rate per 100,000) map visualization was the 42.15 area which shows the highest rate of cancer related deaths in Lake County, Illinois. Another anomaly was that there was a region in the 42.25 area all the way on the right which has a lower rate of cancer related deaths compared to the regions around it which have higher rates.
